```python
ds_hw_grow = dfcsv1.sort_values(["Выпуск", "Data1"])
ds_hw_grow
```

```python
df = ds_hw_grow
n_vyp = df["ВыпускID"].nunique()
n_cols = 10
n_rows = int(np.ceil(n_vyp / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 8))

for ii, vypusk in enumerate(df["ВыпускID"].unique()):
    row, col = divmod(ii, n_cols)
    (df
     .query("ВыпускID == @vypusk")
     .set_index("РейтингID")
     ["Data1"]
     .plot(ax=axes[row, col], sharey=True, title=vypusk, grid=True, fontsize=4))
    axes[row, col].set_xlabel(None)

plt.tight_layout()
```

```python
dfv = pd.DataFrame(columns = ['Выпуск', 'Движение'])
for i in range(n_vyp):
    dfv.loc[i+1] = [i+1, 0]
```

```python
for i in range(dfcsv1.shape[0]-5):
    if dfcsv1['Выпуск'][i] == dfcsv1['Выпуск'][i+1]:
        ggg = 0
        iii = int(dfcsv1['Выпуск'][i])
        if int(dfv['Движение'][iii]) > 0:
            ggg = int(dfv['Движение'][iii])
        dfv['Движение'][iii] = int(ggg + abs(dfcsv1['РейтингID'][i]-dfcsv1['РейтингID'][i+1]))
```
